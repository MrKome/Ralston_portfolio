```python
# This code explores the NBA players from 20xx to 20xx basketball season. Additionally it uses a machine learning algorithm called kMeans to group them in clusters
# As a result, this will show which players are identical.
```

```python
#Import the depencdencies
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
#Load the data
#from google.colab import files
#uploaded =files.upload()
nba = pd.read_csv('nba_2013.csv')
nba.head(7)
```

```python
#Get the number of rows and columns  (481 rows or players and 31 columns or features)
nba.shape
```

```python
#find the average value for each numeric column
nba.mean()
```

```python
# Get the mean or average for a specific column
nba.loc[:,'age'].mean()
```

```python
# Make a pairwise scatter plot (This is a good way to explore a dataset to see how different columns correlate to others)
sns.pairplot(nba[["ast", "fg", "trb"]])
```

```python
# To see correlations
nba[["ast", "fg", "trb"]].corr()
```

```python
# Make a heat map to see correlations
correlation = nba[["ast", "fg", "trb"]].corr()
sns.heatmap(correlation, annot=True)
```

```python
# Make the cluster or players using KMeans

from sklearn.cluster import KMeans # importing KMeans
kmeans_model = KMeans(n_clusters=5, random_state=1) # Creating KMeans model with 5 clusters
good_columns = nba._get_numeric_data().dropna(axis=1) # Get numeric data and remove columns with missing data
kmeans_model.fit(good_columns) # Train the model
labels = kmeans_model.labels_ # Get the labels or cluster label for each player
labels
```

```python
# Plot players by cluster
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca_2 = PCA(2)
plot_columns = pca_2.fit_transform(good_columns)
plt.scatter(x=plot_columns[:,0], y=plot_columns[:,1], c=labels)
plt.show()
```

```python
 # Show Kevin Durant in the data and see what cluster he falls into
 KD = good_columns.loc[ nba['player'] == 'Kevin Durant',: ]

# Show Kevin Durant in the data and see what cluster he falls into
 Bron = good_columns.loc[ nba['player'] == 'LeBron James']

 # Print the players
 print(KD)
 print(Bron)
```

```python
# Converting the data into lists for model prediction
Bron_list = Bron.values.tolist()
KD_list = KD.values.tolist()

# The cluster classification of Lebron James and Kevin Durant
Bron_cluster = kmeans_model.predict(Bron_list) # Lebron belongs to cluster 3 and KD belongs to cluster 3
KD_cluster = kmeans_model.predict(KD_list) # When you run the code - kmeans_model.predict(KD_list) seperately from the created variable name, it returns - array([3], dtype=int32).
                                           # This tells us the cluster Kevin is in (You can do the same for LeBron)

#Print the Cluster label
print(Bron_cluster)
print(KD_cluster)
```

```python
# Predict the number of assists per player from field goals made by using ML model Linear Regression
# First  we need to split our data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(nba[['fg']], nba[['ast']], test_size=0.2, random_state=2) # Splitting the data into 80% training and 20% testing
```

```python
#Create the Linear Regression Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)  # Here we train the model ('fit' is another word for train)
predictions = lr.predict(x_test) # Make predictions on the test data

print(predictions) #Print the predictions
print(y_test) #Print the actual values
```

```python
# Test model: To test it we get the Score. Score returns the coefficient of determination, R^2 of the prediction
lr_confidence = lr.score(x_test, y_test)
print("lr confidence", lr_confidence)
```
